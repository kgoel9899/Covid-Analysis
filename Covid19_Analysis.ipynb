```
total_cases = []
total_recovered = []
total_deaths = []
confirmed = confirmed_df.loc[:, '1/22/20':]
deaths = deaths_df.loc[:, '1/22/20':]
recovered = recovered_df.loc[:, '1/22/20':]
confirmed
```

```
india_cases = []
india_recovered = []
india_deaths = []
india_confirmed_df = confirmed_df.loc[confirmed_df['Country/Region'] == 'India']
india_deaths_df = deaths_df.loc[deaths_df['Country/Region'] == 'India']
india_recovered_df = recovered_df.loc[recovered_df['Country/Region'] == 'India']
iconfirmed = india_confirmed_df.loc[:, '1/22/20':]
ideaths = india_deaths_df.loc[:, '1/22/20':]
irecovered = india_recovered_df.loc[:, '1/22/20':]
iconfirmed
```

```
for i in confirmed:
  total_cases.append(confirmed[i].sum())
  total_deaths.append(deaths[i].sum())
  total_recovered.append(recovered[i].sum())
  india_cases.append(iconfirmed[i].sum())
  india_deaths.append(ideaths[i].sum())
  india_recovered.append(irecovered[i].sum())
```

```
X = np.array([i for i in range(1, len(dates) + 1)]).reshape(-1, 1)
y = np.array([i for i in total_cases]).reshape(-1, 1)
y_1 = np.array([i for i in daily_cases]).reshape(-1, 1)
y_2 = np.array([i for i in india_cases]).reshape(-1, 1)
y_3 = np.array([i for i in india_daily_cases]).reshape(-1, 1)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
future = np.array([i for i in range(1, len(dates) + 11)]).reshape(-1, 1)
```

```
d = {'1': 31, '2': 28, '3': 31, '4': 30, '5': 31, '6': 30, '7': 31, '8': 31, '9': 30, '10': 31, '11': 30, '12': 31}
future_dates = [i for i in dates]
for i in range(10):
  last = future_dates[-1]
  last_sep = last.split('/')
  day = int(last_sep[1]) + 1
  month = last_sep[0]
  year = last_sep[2]
  if day > d[last_sep[0]]:
    month = int(last_sep[0]) + 1
    if int(month) > 12:
      year = int(last_sep[2]) + 1
      month = 1
    day = 1
  new_date = str(month) + '/' + str(day) + '/' + str(year)
  future_dates.append(new_date)
future_dates
```

```
poly_reg = PolynomialFeatures(degree = 10)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
pred = lin_reg.predict(poly_reg.fit_transform(future))
# pred
```

```
pred.tolist()
preds = [i[0] for i in pred]
```

```
plt.figure(figsize = (40, 20))
plt.scatter(X, y, color = "red")
plt.plot(future_dates, preds, linewidth = 5)
plt.title('Total Number of Coronavirus Cases (Polynomial Regression Predictions)', size = 40)
plt.xlabel('Date', size = 40)
plt.ylabel('Number of Cases (x1e6)', size = 40)
plt.xticks(rotation = 90)
plt.xticks(size = 25)
plt.yticks(size = 25)
plt.show()
```

```
plt.figure(figsize = (40, 20))
plt.bar(future_dates, preds, linewidth = 5)
plt.title('Total Number of Coronavirus Cases (Polynomial Regression Predictions)', size = 40)
plt.xlabel('Date', size = 40)
plt.ylabel('Number of Cases (x1e6)', size = 40)
plt.xticks(rotation = 90)
plt.xticks(size = 25)
plt.yticks(size = 25)
plt.show()
```
